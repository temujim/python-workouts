```python
reviews_max = {}

for app in android:
    name = app[0]
    n_reviews = float(app[3])
    
    if name in reviews_max and reviews_max[name] < n_reviews:
        reviews_max[name] = n_reviews
        
    elif name not in reviews_max:
        reviews_max[name] = n_reviews
        
        
        
        
```

### 7. Removing Non-English Apps: Part Two

```python
def eng_apps2(appname):
    counter = 0
    
    for char in appname:
        
        if ord(char) > 127:
            counter += 1
            
            if counter > 3:
                return False
            
    return True
            
```

```python
# DQ Solution:

def eng_apps3(appname):
    counter = 0
    
    for char in appname:
        
        if ord(char) > 127:
            counter += 1
            
    if counter > 3:
        return False
    else:    
        return True
```

```python
def freq_table(dataset, index):
    freq = {}
    freqper = {}
    
    for data in dataset:
        val = data[index]
        
        if val in freq:
            freq[val] += 1
        else:
            freq[val] =1
            
    for data2 in freq:
        freqper[data2] = round((freq[data2]/len(dataset))*100, 2)
        
        
    return freqper
            
            

            
            
######## TEMPLATE CODE ########

def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)

    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
        

#display_table()
    
```
